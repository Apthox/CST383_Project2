```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression
from scipy.stats import zscore
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
```

```python
hour_df = pd.read_csv('../datasets/hour.csv')
del hour_df['dteday']
hour_df.info()
```

```python
ranges = [0, 100, 300, 500, 1000]
def apply(row):
    for i in range(len(ranges)):
        if row['cnt'] < ranges[i]:
            return (i - 1)
        


hour_df['group'] = hour_df.apply(apply, axis=1)
```

```python
colors = ['#ff0000', '#00ff00', '#0000ff', '#000000']
cmap = hour_df.apply(lambda row: colors[int(row['group'])], axis=1)

cmap.unique()
plt.scatter(hour_df['weekday'], hour_df['hr'], c=cmap)
```
